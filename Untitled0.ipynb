```python
import geemap
import os
import geemap.colormaps as cm
```

```python
Map = geemap.Map(center=(0, 0), zoom=2)

modis_dataset = ee.ImageCollection('MODIS/061/MOD13A2') \
    .filterDate('2015-01-01', '2022-12-31') \
    .select('NDVI')

image = modis_dataset.toBands()

palette = cm.palettes.ndvi
ndvi_vis = {
  'min': 0.0,
  'max': 8000.0,
  'palette': palette
}

Map.addLayer(image, {}, 'MODIS NDVI Time-series')
Map.addLayer(image.select(20), ndvi_vis, 'MODIS NDVI VIS')

Map.set_plot_options(add_marker_cluster=True, marker=None)

Map.add_colorbar(ndvi_vis, label="NDVI", layer_name='MODIS NDVI VIS')
Map
```

```python
import numpy as np

def ws(y, r=2, d=3):
    y = np.array(y, dtype=np.float64)
    if y[(y<0)|(y>3000)].size:
        raise ValueError("The range of valid values in the dataset is between 0 and 3000!")
    m = y.size
    E = np.eye(m)
    D = np.diff(E, n=d, axis=0)
    t = E+r*np.dot(D.T, D)
    z = np.dot(np.linalg.inv(t), y.T)
    return z
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/root/points_gpp.csv')

for i in range(df.shape[0]):
    id = df.loc[i, 'id']
    point_gpp = df.iloc[i, 3:]
    try:
        smooth_point_gpp = ws(point_gpp)
    except Exception as e:
        print(e)
        continue
    plt.figure(figsize=(16, 4), constrained_layout=True)
    plt.title("Point {}".format(id))
    plt.plot(np.arange(len(point_gpp)), point_gpp, 'k-', label="Raw GPP")
    plt.plot(np.arange(len(smooth_point_gpp)), smooth_point_gpp, 'r--', label="Whittaker")
    plt.xticks(np.arange(len(point_gpp))[::46], ['2015-01-01', '2016-01-01', '2017-01-01', '2018-01-01', '2019-01-01', '2020-01-01'])
    plt.legend()
```

```python
import numpy as np

def ws(y, r=2, d=3):
    y = np.array(y, dtype=np.float64)
    if y[(y<-2000)|(y>10000)].size:
        raise ValueError("The range of valid values should follow the band!")
    m = y.size
    E = np.eye(m)
    D = np.diff(E, n=d, axis=0)
    t = E+r*np.dot(D.T, D)
    z = np.dot(np.linalg.inv(t), y.T)
    return z
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/Alice/global vegetation cover.csv')

for i in range(df.shape[0]):
    id = df.loc[i, 'id']
    point_NDVI = df.iloc[i, 3:]
    try:
        if (point_NDVI < 0).any():
            plt.figure(figsize=(16, 5), constrained_layout=True)
            plt.title("Point {} (Data with Anomalies)".format(id))
            plt.plot(np.arange(len(point_NDVI)), point_NDVI, 'k-', label="NDVI")
            plt.scatter(np.where(point_NDVI < 0)[0], point_NDVI[point_NDVI < 0], color='r', alpha=1, marker='o', label="NaN")
            plt.xticks(np.arange(len(point_NDVI))[::23], ['2015-01-01','2016-01-01','2017-01-01','2018-01-01','2019-01-01','2020-01-01','2021-01-01','2022-01-01'])
            plt.legend(loc='upper left')
        else:
            smooth_point_NDVI = ws(point_NDVI)
            plt.figure(figsize=(16, 5), constrained_layout=True)
            plt.title("Point {} (Smoothed)".format(id))
            plt.plot(np.arange(len(point_NDVI)), point_NDVI, 'k-', label="Raw NDVI")
            plt.plot(np.arange(len(smooth_point_NDVI)), smooth_point_NDVI, 'r--', label="Whittaker (Smoothed)")
            plt.xticks(np.arange(len(point_NDVI))[::23], ['2015-01-01','2016-01-01','2017-01-01','2018-01-01','2019-01-01','2020-01-01','2021-01-01','2022-01-01'])
            plt.legend()
    except Exception as e:
        print(e)
        continue
```
