```python
import ee
import geemap
import pandas as pd
import matplotlib.pyplot as plt

#  Earth Engine
ee.Initialize()

# 
Map = geemap.Map(center=(0, 0), zoom=2)

#  MOD13A2 
modis_dataset = ee.ImageCollection('MODIS/006/MOD13A2') \
    .filterDate('2015-01-01', '2020-12-31') \
    .select('NDVI')

#  image collection  image.
image = modis_dataset.toBands()

# 
palette = cm.palettes.ndvi
ndvi_vis = {
  'min': 0.0,
  'max': 8000.0,
  'palette': palette
}

# 
Map.addLayer(image, {}, 'MODIS NDVI Time-series')
Map.addLayer(image.select(20), ndvi_vis, 'MODIS NDVI VIS')

# 
Map.set_plot_options(add_marker_cluster=True, marker=None)
#  Jupyter Notebook 
Map.add_colorbar(ndvi_vis, label="NDVI", layer_name='MODIS NDVI VIS')
Map
```

```python
# 
out_dir = '/content'  # 
out_csv = os.path.join(out_dir, 'gfc_points.csv')
Map.extract_values_to_points(out_csv)
```

```python
import numpy as np

def ws(y, r=2, d=3):
    y = np.array(y, dtype=np.float64)
    if y[(y>10000)].size:
        raise ValueError("The range of valid values should below 10000!")
    m = y.size
    E = np.eye(m)
    D = np.diff(E, n=d, axis=0)
    t = E+r*np.dot(D.T, D)
    z = np.dot(np.linalg.inv(t), y.T)
    return z
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/gfc_points.csv')

for i in range(df.shape[0]):
    id = df.loc[i, 'id']
    point_gpp = df.iloc[i, 3:]
    try:
        smooth_point_gpp = ws(point_gpp)
    except Exception as e:
        print(e)
        continue
    plt.figure(figsize=(16, 4), constrained_layout=True)
    plt.title("Point {}".format(id))
    plt.plot(np.arange(len(point_gpp)), point_gpp, 'k-', label="Raw NDVI")
    plt.plot(np.arange(len(smooth_point_gpp)), smooth_point_gpp, 'r--', label="Whittaker")
    plt.xticks(np.arange(len(point_gpp))[::46], ['2015-01-01','2017-01-01','2019-01-01'])
    plt.legend()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#  CSV  DataFrame
df = pd.read_csv('/content/gfc_points.csv')

# 
for i in range(df.shape[0]):
    #  ID
    id = df.loc[i, 'id']
    #  4 
    point_gpp = df.iloc[i, 3:]
    try:
        # 
        if (point_gpp < 0).any():
            # 
            plt.figure(figsize=(16, 4), constrained_layout=True)
            # 
            plt.title("Point {} (Data with Anomalies)".format(id))
            # 
            plt.plot(np.arange(len(point_gpp)), point_gpp, 'k-', label="NDVI")
            # 
            plt.scatter(np.where(point_gpp < 0)[0], point_gpp[point_gpp < 0], color='r', label="Abnormal NDVI")
            #  x 
            plt.xticks(np.arange(len(point_gpp))[::46], ['2015-01-01','2017-01-01','2019-01-01'])
            # 
            plt.legend(loc='upper left')
        else:
            #  Whittaker 
            smooth_point_gpp = ws(point_gpp)
            # 
            plt.figure(figsize=(16, 4), constrained_layout=True)
            # 
            plt.title("Point {} (Smoothed)".format(id))
            # 
            plt.plot(np.arange(len(point_gpp)), point_gpp, 'k-', label="Raw NDVI")
            #  Whittaker 
            plt.plot(np.arange(len(smooth_point_gpp)), smooth_point_gpp, 'r--', label="Whittaker (Smoothed)")
            #  x 
            plt.xticks(np.arange(len(point_gpp))[::46], ['2015-01-01','2017-01-01','2019-01-01'])
            # 
            plt.legend()
    except Exception as e:
        # 
        print(e)
        continue
```
